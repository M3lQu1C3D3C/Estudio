# Editar Imagenes con OPENCV  
Las librerias necesarias serán:

```python
import cv2 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

## 1. Leer una imagen y guardarla

```python
# Leer la imagen
imagen = cv2.imread("inputTeam.jpg")
# Cambiar de color en un lugar 
imagen[:200, :300] = (255, 255, 0)
# Guardar la imagen 
cv2.imwrite("outputTeam.jpg", imagen)
```

## 2. Dibujar figuras en la imagen  

```python
imagen = cv2.imread("inputTeam.jpg")
# Dibujar una línea
cv2.line(imagen, (0, 0), (500, 500), (255, 0 , 0), 5)
# Dibujar un rectángulo
# Dibujar un circulo
# Guargar la imagen 
cv2.imwrite("output_2.jpg", imagen)
```

## Visualizar la imagen editada

```python
img = mpimg.imread("output_2.jpg")
plt.imshow(img)
```
